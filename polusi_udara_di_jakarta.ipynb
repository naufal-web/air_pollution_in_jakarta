Yang salah adalah datanya agak kurang konsisten. Misalnya, data antara "DKI1 Bundaran Hotel Indonesia (HI)" dan "DKI1 Bunderan HI". Namanya berbeda, padahal konteksnya sama. Solusinya adalah menormalkan data tersebut dengan cara yang berbeda. Contoh sebagai berikut.

```python
def normalize_data(data:str):
  if data.startswith("DKI1"):
    return "DKI1 Bundaran Hotel Indonesia"
  elif data.startswith("DKI2"):
    return "DKI2 Kelapa Gading"
  elif data.startswith("DKI3"):
    return "DKI3 Jagakarsa"
  elif data.startswith("DKI4"):
    return "DKI4 Lubang Buaya"
  elif data.startswith("DKI5"):
    return "DKI5 Kebon Jeruk"
  else:
    return
```

Lalu kita uji olah data tersebut dengan fungsi tersebut.

```python
this_not_null_dataframe.stasiun = this_not_null_dataframe.stasiun.apply(normalize_data)
set(this_not_null_dataframe.stasiun.values.tolist())
```

Seperti contoh di atas kita akan buat pernyataan data khusus kurun waktu tertentu

```python
def generate_januari(start:int, stop:int, year:str):
  for day in range(32):
    if start <= day <= stop:
      yield f"{day} Januari {year}"
```

Kita akan coba fungsi tersebut.

```python
for tanggal in generate_januari(24, 31, "2024"):
  print(tanggal)
```

Seperti contoh di atas, kita akan buat fungsi yang sama namun berbeda.

```python
def generate_februari(start:int, stop:int, year:str):
  if int(year) % 4:
    for day in range(30):
      if start <= day <= stop:
        yield f"{day} Februari {year}"
  else:
    for day in range(29):
      if start <= day <= stop:
        yield f"{day} Februari {year}"
```

```python
for tanggal in generate_februari(1, 13, "2024"):
  print(tanggal)
```
