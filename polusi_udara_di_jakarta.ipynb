Seperti contoh di atas kita akan buat pernyataan data khusus kurun waktu tertentu

```python
def generate_januari(start:int, stop:int, year:str):
  for day in range(32):
    if start <= day <= stop:
      yield f"{day} Januari {year}"
```

Kita akan coba fungsi tersebut.

```python
for tanggal in generate_januari(24, 31, "2024"):
  print(tanggal)
```

Seperti contoh di atas, kita akan buat fungsi yang sama namun berbeda.

```python
def generate_februari(start:int, stop:int, year:str):
  if int(year) % 4:
    for day in range(30):
      if start <= day <= stop:
        yield f"{day} Februari {year}"
  else:
    for day in range(29):
      if start <= day <= stop:
        yield f"{day} Februari {year}"
```

```python
for tanggal in generate_februari(1, 13, "2024"):
  print(tanggal)
```

Sederhanakan langkah-langkah tersebut menjadi sebuah fungsi

```python
def transform_data(raw_data: pd.DataFrame, location:str, category: str, period: list[str]):
  processing_data = raw_data.copy()
  processing_data = processing_data[processing_data.tanggal.apply(lambda x: x in period)][(processing_data.stasiun == location) & (processing_data.kategori == category)]
  del raw_data
  raw_data = processing_data
  return raw_data

transform_data(pm_25, "DKI4 Lubang Buaya", "SEDANG", periode)
```
